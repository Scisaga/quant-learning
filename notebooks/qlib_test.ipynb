```python
import qlib
from qlib.data import D
from qlib.config import C
from qlib.constant import REG_CN
from typing import Any, Dict
import pandas as pd

qlib.init(provider_uri="../data/qlib_data/cn_data", region=REG_CN)
inst = "SH600000"  # 换成你的标的
# 选一列稳妥的字段（$close 或 $factor），用很宽的时间窗
s = D.features([inst], ["$close"], "2000-01-01", "2100-01-01", freq="day")["$close"].dropna()

start_dt = s.index.get_level_values("datetime")[0]
end_dt   = s.index.get_level_values("datetime")[-1]
print(inst, "start:", start_dt.date(), "end:", end_dt.date())
```
